```python
import re
import nltk
import numpy as np
import pandas as pd
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
```

```python
file_name = list(uploaded.keys())[0]

with open(file_name, "r", encoding="utf-8") as f:
    cover_letter_text = f.read()

cover_letter_text[:300]
```

```python
cover_letters = re.split(r'_{10,}', cover_letter_text)
cover_letters = [cl.strip() for cl in cover_letters if len(cl.strip()) > 150]

len(cover_letters)
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

def analyze_text(text):
    stop_words = set(stopwords.words('english'))
    lemmatizer = WordNetLemmatizer()

    text = text.lower()
    text = re.sub(r'[^a-z\s]', ' ', text)

    tokens = text.split()
    tokens = [w for w in tokens if w not in stop_words and len(w) > 2]
    tokens = [lemmatizer.lemmatize(w) for w in tokens]

    return " ".join(tokens)
```

```python
analyzed_letters = [analyze_text(cl) for cl in cover_letters]

analyzed_letters[0][:250]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_model = TfidfVectorizer(max_features=1000)
tfidf_matrix = tfidf_model.fit_transform(analyzed_letters)

tfidf_matrix.shape
```

```python
keywords = tfidf_model.get_feature_names_out()
scores = np.asarray(tfidf_matrix.mean(axis=0)).ravel()

keyword_df = pd.DataFrame({
    "Keyword": keywords,
    "Importance": scores
}).sort_values(by="Importance", ascending=False)

keyword_df.head(20)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_scores = cosine_similarity(tfidf_matrix)

similarity_scores
```

```python
def find_similar_letters(index, top_n=3):
    scores = similarity_scores[index]
    similar_indices = scores.argsort()[::-1][1:top_n+1]
    return similar_indices

find_similar_letters(0)
```

```python
import re

file_path = "/content/Data CL (2).txt"  # CORRECT PATH

with open(file_path, "r", encoding="utf-8") as f:
    raw_text = f.read()

print(raw_text[:500])  # preview
```

```python
documents = re.split(r"_+", raw_text)
documents = [doc.strip() for doc in documents if len(doc.strip()) > 200]

print("Total documents:", len(documents))
```

```python
def preprocess_text(text):
    text = text.lower()

    # remove emails
    text = re.sub(r"\S+@\S+", " ", text)

    # remove phone numbers
    text = re.sub(r"\+?\d[\d\s\-()]{7,}", " ", text)

    # remove punctuation & digits
    text = re.sub(r"[^a-z\s]", " ", text)

    # remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    return text
```

```python
cleaned_documents = [preprocess_text(doc) for doc in documents]

print(cleaned_documents[0][:300])
```

```python
output_path = "/content/cleaned_cover_letters.txt"

with open(output_path, "w", encoding="utf-8") as f:
    for doc in cleaned_documents:
        f.write(doc + "\n")

print("Saved cleaned file at:", output_path)
```

```python
import re

jd_file_path = "/content/Job Description.txt"

with open(jd_file_path, "r", encoding="utf-8") as f:
    jd_raw_text = f.read()

print(jd_raw_text[:500])
```

```python
jd_documents = re.split(r"_+", jd_raw_text)

# remove small junk chunks
jd_documents = [doc.strip() for doc in jd_documents if len(doc.strip()) > 150]

print("Total JD documents:", len(jd_documents))
```

```python
def preprocess_text(text):
    text = text.lower()

    # remove emails
    text = re.sub(r"\S+@\S+", " ", text)

    # remove phone numbers
    text = re.sub(r"\+?\d[\d\s\-()]{7,}", " ", text)

    # remove punctuation & digits
    text = re.sub(r"[^a-z\s]", " ", text)

    # normalize spaces
    text = re.sub(r"\s+", " ", text).strip()

    return text
```

```python
cleaned_jd_documents = [preprocess_text(doc) for doc in jd_documents]

print(cleaned_jd_documents[0][:300])
```

```python
jd_output_path = "/content/cleaned_job_descriptions_spaced.txt"

with open(jd_output_path, "w", encoding="utf-8") as f:
    for doc in cleaned_jd_documents:
        f.write(doc + "\n\n")   # ðŸ‘ˆ extra space added

print("Saved spaced JD file at:", jd_output_path)
```
